# ***Importing required libraries***

```python
import numpy as np #numpy is used for linear algebra, fourier transforms
import pandas as pd #pandas is used for working with data, analysing, cleaning, exploring and maipulating of data
```

```python
#read the iris dataset
dframe=pd.read_csv('Iris.csv')
```

```python
#view the dataset as a dataframe
dframe
```

```python
#dimensions of the data set
dframe.shape
```

# ***Cleaning of data***

```python
#in our data set the id column is of no use and irrelevant, so we drop the column
dframe.drop(columns=["Id"],inplace=True)
```

```python
#display 1st five rows in dataset
dframe.head()
```

***we convert the categorica data into numerical data, it improves the machine learning algorithms compatible, increases efficiency of the computation,increases interpretability***

```python
#we assign 1 to Iris-setosa, 2 to Iris-versicolor, 3 to Iris-virginica
dframe['Species'].replace({"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica":3},inplace = True)
dframe
```

***now, we create the arrays of the data set with four attributes, sepal length, sepal width, petallength and petall width***

```python
x=pd.DataFrame(dframe,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
x
```

```python
y=dframe.Species.values.reshape(-1,1)
y
```

***Let us split our data into training and testing set, for this we use scikit learn library in python***

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

***Import the Random Forest Classifier from sklearn***

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100)
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)
```

***we now find the accuracy of our model***

```python
from sklearn import metrics
print("Accuracy: ",metrics.accuracy_score(y_test,y_predict))
```
